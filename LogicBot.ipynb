```python
from kanren import Relation, facts, run, conde, var
import toolz

father = Relation()

facts(
    father,
    ("Vito", "Michael"),
    ("Vito", "Sonny"),
    ("Vito", "Fredo"),
    ("Michael", "Anthony"),
    ("Michael", "Mary"),
    ("Sonny", "Vicent"),
    ("Sonny", "Francesca"),
    ("Sonny", "Kathryn"),
    ("Sonny", "Frank"),
    ("Sonny", "Santino"),
)


def buscar_hijos(papa):
  return((run(0, q, father(papa, q))))

def agregar_vinculo_padre(a,b):
  facts(father,(a,b))

def agregar_vinculo_hijo(a,b):
  facts(father,(b,a))

def agregar_vinculo_abuelo(sujeto, abuelo): #Agregar un papá al papá del sujeto, el sujeto DEBE de existir en la base de conocimientos
  q = var()
  valor = (run(0, q, father(q, sujeto)))
  padre = str(valor[0])
  facts(father,(abuelo, padre))
  return run(0, q, father(q, padre))

def agregar_vinculo_hermano(sujeto, hermano): #agregar un hijo a los padres del sujeto
  q = var()
  valor1 = (run(0, q, father(q, sujeto)))
  padre = str(valor1[0])
  facts(father,(padre, hermano))
  return run(0, q, father(padre, q))

def agregar_vinculo_tio(sujeto, tio): #agregar hermano al papá
  q = var()
  valor1 = (run(0, q, father(q, sujeto)))
  padre = str(valor1[0])
  agregar_vinculo_hermano(padre,tio)
  return

def agregar_vinculo_primo(sujeto, primo): #Agregar un hijo al hermano del papá del sujeto
  q = var()
  valor1 = (run(0, q, father(q, sujeto)))
  padre = str(valor1[0])
  valor2 = buscar_sibling(padre)
  tio = str(valor2[0])
  facts(father,(tio, primo))
  return run(0, q, father(tio, q))

def buscar_grandparent(gparent, child):
    p = var()
    return conde((father(gparent, p), father(p, child)))

def buscar_sibling(hermano):
    p = var()
    b = var()
    res = conde((father(p, hermano), father(p, b))) # Buscas a "b" que sea QUIEN COMPARTA LOS MISMOS PADRES QUE "hermano"
    resultado = list(run(0, b, res, results_filter=toolz.unique))
    resultado.remove(hermano)
    return resultado

def buscar_tios(persona):
    x = var()
    y = var()
    z = var()

    metas = conde((father(x, y), father(y, persona), father(x, z)))
    resultados =list( run(0, z, metas, results_filter=toolz.unique))
    papa = list (run(0, x, father(x, persona)))
    lista_resultante = [elemento for elemento in resultados if elemento not in papa]
    return lista_resultante

def buscar_primos(persona):
    x = var()
    y = var()
    z = var()
    p = var()

    metas = conde((father(x, y), father(y, persona), father(x, z), father(z, p)))
    resultados = list( run(0, p, metas, results_filter=toolz.unique))

    hermanos = buscar_sibling(persona)
    lista_resultante = [elemento for elemento in resultados if elemento not in hermanos]
    lista_resultante.remove(persona)
    return lista_resultante

q = var()

while True:
    opcion = input("Ingresa una opción (1: Añadir vínculo familiar,2: Buscar Padres: ,3: Buscar Abuelos, , 4: Buscar Hermanos, 5: Buscar Tios, 6: Buscar Primos:,7: Mostrar toda la base de conocimientos , 8: Salir): ")
    if opcion == "1":
        persona = input("\n (Para registrarte en la base de conocimientos debes 1ro añadir un vinculo padre e hijo) \nIngresa tu Nombre: ")
        while True:
          option = input("Ingresa una opción (1: Añadir padres, 2: Añadir hijo, 3: Añadir Abuelo, 4: Añadir hermano, 5: Añadir Tio, 6: Añadir primo, 7: Volver al menu anterior)")
          if option == "1":
            papa = input("Ingresa el nombre del padre: ")
            agregar_vinculo_padre(papa,persona)
          elif option == "2":
            a = input("Ingresa el nombre del hijo: ")
            agregar_vinculo_hijo(a,persona)
          elif option == "3":
            a = persona
            b = input("Ingresa el nombre del abuelo: ")
            agregar_vinculo_abuelo(a,b)
          elif option == "4":
            a = persona
            b = input("Ingresa el nombre del hermano: ")
            agregar_vinculo_hermano(a,b)
          elif option == "5":
            a = persona
            print("ADVERTENCIA: Para esta opcion tienes que tener un abuelo\n")
            b = input("Ingresa el nombre del Tio: ")
            agregar_vinculo_tio(a,b)
          elif option == "6":
            a = persona
            print("ADVERTENCIA: Para esta opcion tienes que tener un tio\n")
            b = input("Ingresa el nombre del primo: ")
            agregar_vinculo_primo(a,b)
          elif option == "7":
            break
          else:
            print("Opción inválida. Intenta nuevamente.")
    elif opcion == "2":
        a = input("Ingresa el nombre de a: ")
        print(run(0, q, father(q,a)))
    elif opcion == "3":
        a = input("Ingresa el nombre de a: ")
        print( run(0, q, buscar_grandparent(q,a)) )
    elif opcion == "4":
        a = input("Ingresa el nombre de a: ")
        print(buscar_sibling(a))
    elif opcion == "5":
        a = input("Ingresa el nombre de a: ")
        print (buscar_tios(a))
    elif opcion == '6':
        a = input("Ingresa el nombre de a: ")
        print(buscar_primos(a))
    elif opcion == '7':
        print("\n" ,father.facts, "\n")
    elif opcion == '8':
        break
    else:
        print("Opción inválida. Intenta nuevamente.")
```
